1. Difference between a function and a method in Python

* Function: A block of reusable code that performs a specific task, defined using
def.

* Method: A function that is associated with an object and is called using dot notation.

* Example Function

      def greet(name):
          return f"Hello, {name}!"

* Example Method

      text = "hello"
      print(text.upper())   # upper() is a method of string objects

2. Function arguments and parameters in Python
* Parameters: Variables listed inside parentheses in the function definition.

* Arguments: The actual values passed to the function when it is called.

* example

      def add(a, b):   # a and b are parameters
          return a + b

      print(add(3, 5))  # 3 and 5 are arguments

3. Ways to define and call a function

* Normal function with def

* Lambda function

* Recursive function

* example

      # Normal
      def square(x):
          return x*x
      print(square(4))

      # Lambda
      sq = lambda x: x*x
      print(sq(4))

      # Recursive
      def factorial(n):
          return 1 if n==0 else n*factorial(n-1)
      print(factorial(5))


4. Purpose of return

* Sends a value back to the caller.

* Ends function execution.

* example

      def multiply(a, b):
          return a * b

      result = multiply(3, 4)
      print(result)   # 12

5. Iterators vs Iterables

* Iterable: Any object that can return an iterator (e.g. list, tuple, string).

* Iterator: An object that produces items one at a time using __next__().

* example

      nums = [1, 2, 3]    # iterable
      it = iter(nums)     # iterator
      print(next(it))     # 1
      print(next(it))     # 2

6. Concept of Generators

* Special functions using yield instead of return.

* Generate values lazily (one at a time).

* example

      def countdown(n):
          while n > 0:
              yield n
              n -= 1

      for i in countdown(3):
          print(i)

7. Advantages of generators

* Memory efficient (don’t store all values).

* Useful for large datasets or infinite sequences.

* Lazy evaluation.

8. Lambda functions
* Anonymous, single-line functions using lambda.

* Typically used in map(), filter(), sorted().

      square = lambda x: x*x
      print(square(5))

9. Purpose of map()

* Applies a function to every element of an iterable.

* example

      nums = [1, 2, 3]
      doubled = list(map(lambda x: x*2, nums))
      print(doubled)  # [2, 4, 6]

10. Difference between map(), reduce(), filter()

* map() → applies a function to all elements → returns list/iterator.

* filter() → selects elements based on condition.

* reduce() → applies function cumulatively to reduce sequence to single value.

* example

      from functools import reduce

      nums = [1, 2, 3, 4]

      print(list(map(lambda x: x*2, nums)))      # [2, 4, 6, 8]
      print(list(filter(lambda x: x%2==0, nums))) # [2, 4]
      print(reduce(lambda x,y: x+y, nums))        # 10

```python
# 1. Sum of even numbers

def sum_even(lst):
    return sum(x for x in lst if x % 2 == 0)

print(sum_even([1,2,3,4,5,6]))  # 12
```

```python
# 2. Reverse a string

def reverse_string(s):
    return s[::-1]

print(reverse_string("Python"))  # nohtyP
```

```python
# 3. Squares of numbers

def squares(lst):
    return [x**2 for x in lst]

print(squares([1,2,3,4]))  # [1, 4, 9, 16]
```

```python
# 4. Prime number check (1–200)

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

primes = [x for x in range(1,201) if is_prime(x)]
print(primes)
```

```python
# 5. Fibonacci Iterator

class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a, self.b, self.count = 0, 1, 0
    def __iter__(self):
        return self
    def __next__(self):
        if self.count >= self.n:
            raise StopIteration
        self.count += 1
        self.a, self.b = self.b, self.a + self.b
        return self.a

for num in Fibonacci(10):
    print(num)
```

```python
# 6. Powers of 2 generator

def powers_of_two(n):
    for i in range(n+1):
        yield 2**i

print(list(powers_of_two(5)))
```

```python
# 7. File line generator

from google.colab import files

# Step 1: Upload the file
uploaded = files.upload()

# Step 2: Define generator
def read_file_lines(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            yield line.rstrip('\n')

# Step 3: Automatically detect the uploaded filename
for filename in uploaded.keys():
    print(f"\nReading from {filename}:\n")
    for line in read_file_lines(filename):
        print(line)
```

```python
# 8. Sort tuples by second element

data = [(1,3), (2,1), (5,2)]
sorted_list = sorted(data, key=lambda x: x[1])
print(sorted_list)  # [(2,1), (5,2), (1,3)]
```

```python
# 9. Celsius → Fahrenheit using map()

celsius = [0, 10, 20, 30]
fahrenheit = list(map(lambda c: (c*9/5)+32, celsius))
print(fahrenheit)  # [32, 50, 68, 86]
```

```python
# 10. Remove vowels with filter()

s = "Hello World"
result = ''.join(filter(lambda x: x.lower() not in 'aeiou', s))
print(result)  # Hll Wrld
```
